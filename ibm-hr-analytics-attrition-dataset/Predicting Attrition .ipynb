```python
df['EducationField']=encode.fit_transform(df['EducationField'].astype('str'))
```

```python
df['Gender']=encode.fit_transform(df['Gender'].astype('str'))
```

```python
df.drop('JobRole',axis=1,inplace=True)
```

```python
df['MaritalStatus']=encode.fit_transform(df['MaritalStatus'].astype('str'))
```

```python
df['Over18']=encode.fit_transform(df['Over18'].astype('str'))
```

```python
df['OverTime']=encode.fit_transform(df['OverTime'].astype('str'))
```

```python
def converter(attrition):
    if attrition == 'Yes':
        return 1
    else :
        return 0
```

```python
df['Attrition1']=df['Attrition'].apply(converter)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'C':[1,0.1,10,12,13,23,56,100,145,200,300,1000,0.01],'gamma':[1,0,0.1,0.5,0.75,0.01,0.001,0.0001,10]}
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
grid=GridSearchCV(SVC(),param_grid,verbose=3,refit=True)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
pred_i = grid.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred_i))
print('\n')
print(classification_report(y_test,pred_i))
```

```python
#This is a good model, but the prediction for 1 is low, so we cant use Support Vector Machines
#I am thinking this problem arises due to low number of samples of prediction 'Yes' as compared to no
#Then we have to use SMOTE
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE()
```

```python
X_sm,y_sm=smote.fit_sample(X,y)
```

```python
X_train_sm,X_test_sm,y_train_sm,y_test_sm=train_test_split(X_sm,y_sm,test_size=0.3,random_state=101)
```

```python
grid.fit(X_train_sm,y_train_sm)
```

```python
grid.best_params_
```

```python
pred=grid.predict(X_test_sm)
```

```python
print(confusion_matrix(y_test_sm,pred))
print('\n')
print(classification_report(y_test_sm,pred))
```
